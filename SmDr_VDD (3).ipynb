## Данил

```python
df = df[df['SMK_stat_type_cd'] != 2]
df['SMK_stat_type_cd'] = df['SMK_stat_type_cd'].map({1.0: 0, 3.0: 1})

correlation_matrix = df.corr()

target_corr = correlation_matrix['SMK_stat_type_cd'].sort_values(ascending=False)
print(target_corr)
```

```python
columns_to_drop = ['BLDS', 'SBP', 'sight_right', 'sight_left', 'SGOT_AST', 
                   'urine_protein', 'tot_chole', 'LDL_chole', 'hear_left', 
                   'hear_right', 'age', 'HDL_chole']
df = df.drop(columns=columns_to_drop)
```

```python
from sklearn.utils import resample

# Отделяем классы
class_1 = df[df['SMK_stat_type_cd'] == 0]
class_2 = df[df['SMK_stat_type_cd'] == 1]

# Уменьшаем больший класс до размера меньшего
class_1_downsampled = resample(class_1,
                               replace=False,  # Уменьшение
                               n_samples=len(class_2),
                               random_state=42)

# Объединяем данные обратно
df = pd.concat([class_1_downsampled, class_2])

# Проверяем распределение классов
print(df['SMK_stat_type_cd'].value_counts())
```

```python
X = df.drop(columns=['SMK_stat_type_cd'])
y = df['SMK_stat_type_cd']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Размер обучающей выборки:", X_train.shape)
print("Размер тестовой выборки:", X_test.shape)
```

# Логистическая регрессия

```python
model1 = LogisticRegression(max_iter=1000, random_state=42)
model1.fit(X_train, y_train)
```

```python
y_pred1 = model1.predict(X_test)
accuracy1 = model1.score(X_test, y_test)
class_report1 = classification_report(y_test, y_pred1)
conf_matrix1 = confusion_matrix(y_test, y_pred1)
```

# Ближайшие соседи

```python
model2 = KNeighborsClassifier(n_neighbors=5)
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_test)
accuracy2 = model2.score(X_test, y_test)
class_report2 = classification_report(y_test, y_pred2)
conf_matrix2 = confusion_matrix(y_test, y_pred2)
```

# Случайные деревья

```python
model3 = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42)
model3.fit(X_train, y_train)
```

```python
y_pred3 = model3.predict(X_test)
accuracy3 = model3.score(X_test, y_test)
class_report3 = classification_report(y_test, y_pred3)
conf_matrix3 = confusion_matrix(y_test, y_pred3)
```

```python
classifiers =["Logistic Regression", "KNeighborsClassifier", "Random Forest Classifier"]

log_cols = ["Classifier", "Accuracy"]
log = pd.DataFrame(columns=log_cols)

log["Classifier"] = classifiers
log["Accuracy"] = [accuracy1, accuracy2, accuracy3]
log
```
