```python
from sklearn.model_selection import train_test_split
X = pd.concat([train[feat]], axis=1)
X.shape
```

```python
X.head(5).T
```

```python
y = train["price_range"]
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.metrics import accuracy_score
```

### ближайшие соседи

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
%%time
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
knn_pred_test = knn.predict(X_test)
display(knn_pred_test[:20], np.array(y_test[:20]))
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

print('train accuracy: ', accuracy_score(y_train, y_train_predict, normalize=True))
print('test accuracy: ', accuracy_score(y_test, y_test_predict, normalize=True))
```

```python
knn_accuracy = accuracy_score(y_test, y_test_predict, normalize=True)
```

```python
print(classification_report( y_test, y_test_predict))
```

```python
gnb_accuracy = accuracy_score(y_test, gnb_pred_test)
```

### Дерево решений

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
%%time
rfc  = RandomForestClassifier(criterion = 'gini',
                                    n_estimators = 100,
                                    max_depth = 9,
                                    random_state = 42,
                                    n_jobs = -1)

rfc.fit(X_train, y_train)
rfc_y_test_pred = rfc.predict(X_test)
rfc_y_train_predict = rfc.predict(X_train)
```

```python
rfc_pred_train = rfc.predict(X_train)
rfc_pred_test = rfc.predict(X_test)
display(rfc_pred_test[:20], np.array(y_test[:20]))
```

```python
from sklearn.metrics import accuracy_score
```

```python
rfc_train_accuracy_score = accuracy_score(y_train, rfc_pred_train, normalize=True)
rfc_test_accuracy_score = accuracy_score(y_test, rfc_pred_test, normalize=True)

print('train accuracy: ', rfc_train_accuracy_score)
print('test accuracy: ', rfc_test_accuracy_score)
```

```python
rfc_accuracy = accuracy_score(y_train, rfc_pred_train, normalize=True)
```

### Logistic Regression (Логистическая регрессия)

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc_X_train = sc.fit_transform(X_train)
sc_X_test = sc.transform(X_test)
```

```python
%%time
lr = LogisticRegression(random_state = 42)
lr.fit(sc_X_train, y_train)
```

```python
sc_X_train
```

```python
sc_X_test
```

```python
lr_pred_train = lr.predict(X_train)
lr_pred_test = lr.predict(X_test)
display(lr_pred_test[:20], np.array(y_test[:20]))
```

```python
from sklearn.metrics import accuracy_score
lr_train_accuracy_score = accuracy_score(y_train, lr_pred_train)
lr_test_accuracy_score = accuracy_score(y_test, lr_pred_test)

print('train accuracy: ', lr_train_accuracy_score)
print('test accuracy: ', lr_test_accuracy_score)
```

```python
lr_accuracy = accuracy_score(y_test, lr_pred_test)
```

```python
print(classification_report( y_test, y_test_predict))
```
