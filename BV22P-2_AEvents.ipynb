### Предобработка данных

```python
ae.duplicated().sum()
```

```python
# проверка на отсутствующие значения для всех колонок датафрейма
ae.isna().sum()
```

```python
ae['Medal'].value_counts()
```

```python
ae['Medal'].mode()[0]
```

```python
ae['Medal'] = ae['Medal'].fillna(value=ae['Medal'].mode()[0])
```

```python
ae.isna().sum()
```

```python
ae['Height'] = ae['Height'].fillna(value=ae['Height'].mode()[0])
ae['Weight'] = ae['Weight'].fillna(value=ae['Weight'].mode()[0])
ae['Age'] = ae['Age'].fillna(value=ae['Age'].mode()[0])
```

```python
ae.isna().sum()
```
